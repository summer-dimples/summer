```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
v_max = 70  # Maximum velocity
rho_max = 100  # Maximum density

# Define density values
rho = np.linspace(0, rho_max, 100)
# Compute velocity using the nonlinear speed-density relationship
v = v_max * (1 - (rho / rho_max) ** 2)

plt.figure(figsize=(7, 5), dpi=300)
plt.plot(rho, v, 'k-', linewidth=2, label=r"$v = v_{max} \left(1 - \left(\frac{\rho}{\rho_{max}}\right)^2\right)$")

plt.scatter([0, rho_max], [v_max, 0], color='red', zorder=3, s=80, marker='o')

plt.text(5, v_max + 3, r"$v_0$", color='black', fontsize=16, verticalalignment='bottom', horizontalalignment='left')
plt.text(-5, 3, r"$\rho_{min} = 0$", color='black', fontsize=16, verticalalignment='top', horizontalalignment='right')
plt.text(rho_max , 8, r"$\rho_{max}$", color='black', fontsize=16, verticalalignment='bottom', horizontalalignment='left')


plt.xlabel(r'Traffic Density $\rho$', fontsize=16)
plt.ylabel(r'Traffic Velocity $V(\rho)$', fontsize=16)
plt.title("Nonlinear Velocity-Density Relationship", fontsize=16)
plt.xlim(0, rho_max + 10)  
plt.ylim(0, v_max + 10)  
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(fontsize=16)

# Show plot
plt.show()
```

```python
# Define parameters for density-flow relationship
v_max = 70  # Maximum velocity
rho_max = 100  # Maximum density

# Define density values
rho = np.linspace(0, rho_max, 100)

# Compute traffic flow using the formula
q = rho * v_max * (1 - (rho / rho_max) ** 2)

# Compute optimal flow point
rho_opt = rho_max / np.sqrt(3)  # Optimal density
q_opt = (2 / (3 * np.sqrt(3))) * v_max * rho_max  # Optimal flow

plt.figure(figsize=(7, 5), dpi=300)
plt.plot(rho, q, 'k-', linewidth=2, label=r"$q(\rho) = v_{max} \rho \left(1 - \left(\frac{\rho}{\rho_{max}}\right)^2\right)$")

plt.scatter([rho_opt], [q_opt], color='red', zorder=3, s=80, marker='o')
plt.text(rho_opt + 20, q_opt, r"$\mathbf{\left(\frac{\rho_{max}}{\sqrt{3}}, \frac{2}{3\sqrt{3}} v_{max} \rho_{max}\right)}$",
         color='black', fontsize=14, fontweight='bold', verticalalignment='center', horizontalalignment='left')
plt.text(-8, -50, r"$\rho_{min} = 0$", color='black', fontsize=16, verticalalignment='top', horizontalalignment='right')
plt.text(rho_max + 5, -50, r"$\rho_{max}$", color='black', fontsize=16, verticalalignment='top', horizontalalignment='left')
plt.xlabel(r'Traffic Density $\rho$', fontsize=16)
plt.ylabel(r'Traffic Flow $q(\rho)$', fontsize=16)
plt.title("Density-Flow Relationship", fontsize=16)
plt.xticks([0, rho_opt, rho_max], [r"$\rho_{min} = 0$", r"$\frac{\rho_{max}}{\sqrt{3}}$", r"$\rho_{max}$"],fontsize=14)
plt.yticks([0, q_opt], [r"$0$", r"$\frac{2}{3\sqrt{3}} v_{max} \rho_{max}$"],fontsize=14)

plt.xlim(0, rho_max + 20)  
plt.ylim(0, q_opt + 200) 
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(fontsize=16, loc="lower right")

# Show plot
plt.show()
```

```python
# Parameters
Nx = 100  # Number of spatial grid points
Nt = 200  # Number of time steps
L = 1000  # Total road length (m)
T = 50  # Total simulation time (s)
dx = L / Nx  # Spatial step size
dt = T / Nt  # Time step size
v_max = 30  # Maximum speed (m/s)
rho_max = 1.0  # Maximum density (vehicles/m)
rho_left = 0.5  # Initial high density on the left side
rho_right = 0.1  # Initial low density on the right side

# Compute CFL number
cfl = v_max * dt / dx
if cfl > 1:
    raise ValueError("CFL condition violated! Reduce dt or increase dx.")

# Initialize traffic density
rho = np.ones(Nx) * rho_right
rho[:Nx // 2] = rho_left  
rho_new = np.copy(rho)

# Velocity-Density relationship (Greenshields model)
def velocity(rho):
    return v_max * (1 - rho / rho_max)

# Traffic flow function
def flux(rho):
    return rho * velocity(rho)

# Store evolution history
history = [np.copy(rho)]

# Time evolution loop
for n in range(Nt):
    for i in range(1, Nx - 1):
        rho_new[i] = rho[i] - dt / dx * (flux(rho[i]) - flux(rho[i - 1]))
    rho[:] = rho_new
    history.append(np.copy(rho))

plt.figure(figsize=(10, 6))
colors = ['tab:orange', 'tab:red', 'tab:pink', 'tab:blue', 'tab:purple']
for idx, t in enumerate(range(0, Nt, Nt // 5)):
    plt.plot(np.linspace(0, L, Nx), history[t], label=f'Time {t * dt:.1f}s', color=colors[idx], linewidth=2)


plt.xlabel("Position x (m)", fontsize=14)
plt.ylabel("Traffic Density  (vehicles/m)", fontsize=14)
plt.title("Traffic Density Evolution Over Time", fontsize=16, fontweight='bold')
plt.legend(fontsize=16, loc='upper right', edgecolor='black', facecolor='white')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Show plot
plt.show()
```

```python
Nx = 100        # Number of spatial grid points
Nt = 200        # Number of time steps
L = 10.0        # 10 km highway
T = 3.0         # Simulation time in minutes
rho_max = 100   # Updated maximum density (cars/km)
v_max = 70.0    # Updated maximum speed (km/h)
m = 2           # Nonlinear exponent

dx = L / Nx     # Spatial step
dt = T / Nt     # Time step

# CFL condition for stability
dt = min(dt, dx / v_max)

# Initialize density distribution(sine wave disturbance)
def initial_density(x):
    return 50 + 10 * np.sin(2 * np.pi * x / L)  

# Velocity-density relationship
def velocity(rho):
    return v_max * (1 - (rho / rho_max) ** m)

# Compute flux function
def flux(rho):
    return rho * velocity(rho)

# Initialize variables
x = np.linspace(0, L, Nx)
rho = initial_density(x)
rho_new = np.zeros(Nx)

rho_history = []
v_history = []
q_history = []

# Time-stepping loop
for t in range(Nt):
    q = flux(rho)
    
    # Lax-Friedrichs method update
    rho_new[1:-1] = 0.5 * (rho[:-2] + rho[2:]) - (dt / (2 * dx)) * (q[2:] - q[:-2])
    
    # Apply periodic boundary conditions
    rho_new[0] = rho_new[-2]  # Left boundary
    rho_new[-1] = rho_new[1]  # Right boundary
    

    if t in [int(Nt/3), int(2*Nt/3), Nt-1]:
        rho_history.append(rho.copy())
        v_history.append(velocity(rho))
        q_history.append(flux(rho))
    rho[:] = rho_new

rho_values = np.linspace(10, rho_max, 100)
v_values = velocity(rho_values)
plt.figure(figsize=(8, 5))
plt.plot(rho_values, v_values, color='r')
plt.xlabel(r'Traffic Density $\rho$', fontsize=12)
plt.ylabel(r'Traffic Velocity $V(\rho)$', fontsize=12)
plt.title('Non-linear Velocity-Density Relationship', fontsize=12)
plt.grid()
plt.show()

q_values = flux(rho_values)
plt.figure(figsize=(8, 5))
plt.plot(rho_values, q_values, color='b')
plt.xlabel(r'Traffic Density $\rho$', fontsize=12)
plt.ylabel(r'Traffic Flow $q(\rho)$', fontsize=12)
plt.title('Density-Flow Relationship', fontsize=12)
plt.grid()
plt.show()
```
